```python
import sys
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

project_root = Path.cwd().parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 4)
```

```python
SYMBOL = 'VNM'
processed_path = project_root / 'data' / 'processed' / f'{SYMBOL}_processed.csv'
print('Project root:', project_root)
print('Processed path:', processed_path)
print('Exists:', processed_path.exists())
```

```python
df = pd.read_csv(processed_path)
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date').reset_index(drop=True)
df.head()
```

```python
df.describe(include='all').transpose()
```

```python
plt.figure(figsize=(12, 4))
plt.plot(df['date'], df['close'], color='tab:blue')
plt.title(f'Close Price Over Time - {SYMBOL}')
plt.xlabel('Date')
plt.ylabel('Scaled Close Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 4))
plt.bar(df['date'], df['volume'], color='tab:orange')
plt.title(f'Volume Over Time - {SYMBOL}')
plt.xlabel('Date')
plt.ylabel('Scaled Volume')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
tmp = df[['date', 'close']].copy().set_index('date')
decomp = seasonal_decompose(tmp['close'], model='additive', period=30)
fig = decomp.plot()
fig.set_size_inches(12, 8)
plt.tight_layout()
plt.show()
```

```python
num_cols = df.select_dtypes(include='number').columns
plt.figure(figsize=(12, 8))
sns.heatmap(df[num_cols].corr(), cmap='coolwarm', annot=False)
plt.title(f'Correlation Heatmap - {SYMBOL}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['close'], kde=True, color='tab:green')
plt.title(f'Distribution of Close Price - {SYMBOL}')
plt.xlabel('Scaled Close Price')
plt.tight_layout()
plt.show()
```
