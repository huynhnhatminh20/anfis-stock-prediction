```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

from src.data.data_pipeline import StockDataPipeline
```

```python
SYMBOL = 'VNM'
START = '2020-01-01'
END = '2025-12-31'

pipeline = StockDataPipeline(source='VCI')
output = pipeline.fit_transform(symbol=SYMBOL, start=START, end=END)
df = output.full.copy()
df.head()
```

```python
plt.figure(figsize=(12, 4))
plt.plot(df['date'], df['close'])
plt.title('Close Price Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 4))
plt.bar(df['date'], df['volume'])
plt.title('Volume Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
tmp = df.copy().set_index('date')
decomp = seasonal_decompose(tmp['close'], model='additive', period=30)
fig = decomp.plot()
fig.set_size_inches(12, 8)
plt.tight_layout()
```

```python
num_cols = df.select_dtypes(include='number').columns
plt.figure(figsize=(10, 8))
sns.heatmap(df[num_cols].corr(), cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['close'], kde=True)
plt.title('Distribution of Close Price')
plt.tight_layout()
plt.show()
```
